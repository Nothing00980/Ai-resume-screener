```python
# next we will proceed for json text into semantic text

import json

with open("data/description/job_dataset.json", "r", encoding="utf-8") as f:
    jobs_data = json.load(f)

job_texts = []
job_ids = []
job_titles = []

for job in jobs_data:
    print(job.keys())
    combined_text = (
        f"Job Title: {job['Title']}. "
        f"Experience Level: {job['ExperienceLevel']}. "
        f"Years of Experience: {job['YearsOfExperience']}. "
        f"Skills required: {' '.join(job['Skills'])}. "
        f"Responsibilities include: {' '.join(job['Responsibilities'])}. "
        f"Keywords: {' '.join(job['Keywords'])}."
    )
    
    job_texts.append(combined_text)
    job_ids.append(job["JobID"])
    job_titles.append(job["Title"])

print("Total jobs:", len(job_texts))
print("\nSample:\n", job_texts[0][:400])
```

```python
import re

def clean_text(text):
    text = re.sub(r"\S+@\S+", "", str(text))
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text.lower().strip()

resume_texts = [clean_text(t) for t in resume_texts]
job_texts = [clean_text(t) for t in job_texts]

print("Cleaning done.")
print(resume_texts[0][:300])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

vectorizer = TfidfVectorizer(max_features=5000)

tfidf_resumes = vectorizer.fit_transform(resume_texts)
tfidf_jobs = vectorizer.transform(job_texts)

tfidf_similarity = cosine_similarity(tfidf_resumes, tfidf_jobs)

print("TF-IDF similarity matrix shape:", tfidf_similarity)
```

```python
# bert embeddings
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
resume_embeddings = model.encode(
    resume_texts,
    convert_to_tensor=True,
    show_progress_bar=True
)

job_embeddings = model.encode(
    job_texts,
    convert_to_tensor=True,
    show_progress_bar=True
)

print("Resume embedding shape:", resume_embeddings.shape)
print("Job embedding shape:", job_embeddings.shape)
```

```python
bert_similarity = cosine_similarity(
    resume_embeddings.cpu().numpy(),
    job_embeddings.cpu().numpy()
)

print("BERT similarity matrix shape:", bert_similarity.shape)
```

```python
import numpy as np

top_indices = np.argmax(bert_similarity, axis=1)

for i in range(5):  # show first 5
    print("Resume:", resume_names[i])
    print("Matched Job ID:", job_ids[top_indices[i]])
    print("Matched Job Title:", job_titles[top_indices[i]])
    print("Similarity Score:", bert_similarity[i][top_indices[i]])
    print("-----")
```

```python
# ranking metrics
def top_k_accuracy(similarity_matrix, true_indices, k=3):
    top_k = np.argsort(-similarity_matrix, axis=1)[:, :k]
    correct = 0
    for i in range(len(true_indices)):
        if true_indices[i] in top_k[i]:
            correct += 1
    return correct / len(true_indices)
```

```python
def mean_reciprocal_rank(similarity_matrix, true_indices):
    ranks = []
    sorted_indices = np.argsort(-similarity_matrix, axis=1)
    for i in range(len(true_indices)):
        rank = np.where(sorted_indices[i] == true_indices[i])[0][0] + 1
        ranks.append(1/rank)
    return np.mean(ranks)
```

```python
top_k_acc = top_k_accuracy(bert_similarity, top_indices, k=3)
mrr = mean_reciprocal_rank(bert_similarity, top_indices)
```

```python
import numpy as np

top_k = 5
top_k_indices = np.argsort(-bert_similarity, axis=1)[:, :top_k]
```

```python
resume_index = 0

top5_indices_for_resume = top_k_indices[resume_index]
top5_jobs = [job_texts[i] for i in top5_indices_for_resume]

print("Top 5 candidate jobs selected from BERT:")
for i in top5_indices_for_resume:
    print(job_ids[i], "-", job_titles[i])
```

```python
# cross encoder for fine grained re-ranking
from sentence_transformers import CrossEncoder

cross_encoder = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")

resume_text = resume_texts[resume_index]
pairs = [(resume_text, job_text) for job_text in top5_jobs]
scores = cross_encoder.predict(pairs)
```

```python
# re-ranking using cross encoder scores
reranked = list(zip(top5_indices_for_resume, scores))
reranked.sort(key=lambda x: x[1], reverse=True)

print("\nAfter Cross-Encoder Reranking:\n")

for idx, score in reranked:
    print("Job:", job_ids[idx], "-", job_titles[idx])
    print("Score:", score)
    print("------")
```
